```python

import csv
import re
f = open("data/dataSet.csv", encoding="cp949")
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
header2 = data[0:3]

del data[0:3]
del data[270]

for i in range(len(data)):
    for j in range(2,len(data[0])):
        data[i][j] = re.sub(",","",data[i][j])


for row in data:
    print(row)
```

```python
for row in data:
    print(row)
```

```python
def CountNan(data, col):
    count = 0
    for row in data:
        if row[col] == "-":
            count += 1
    return count

length = len(data[0])

for col in range(length-1, 1, -1):
    nan_count = CountNan(data, col)
    if nan_count <= 108:
        for i in range(len(data)):
            if data[i][col] == "-":
                data[i][col] = '0'
    elif nan_count > 108:
        for i in range(len(data)):
            data[i].pop(col)
        for row in header2:
            row.pop(col)
```

```python
length = len(data[0])
x_val=[]
local = []
for x in range(int((len(header2[1])-2)/3)):
    x_val.append(header2[1][x*3+2])

for row in data:
    if row[0].split("지방경찰청")[0] not in local:
        local.append(row[0].split("지방경찰청")[0])
```

```python
def plotdata(data,loc):
    sum = [0 for i in range(int((length-2)/3))]
    for  row in data:
        if row[0].split('지방경찰청')[0] == loc:
            for col in range(0,int((length-2)/3)):
                sum[col] += int(row[3*col+2])

    return sum,loc

n = len(local)
r = 6
c = (n + 1) // r 
fig, axes = plt.subplots(r, c, figsize=(5*c, 3*r),constrained_layout=True) 
axes = axes.flatten()
for i, x in enumerate(local):
    sum,loc = plotdata(data,x)
    axes[i].bar(x_val,sum,width=0.8,color = "#003458")
    axes[i].set_title(loc)
    axes[i].set_ylabel('사고(건)')
    axes[i].set_xticklabels(x_val, rotation=45, ha='right')
plt.tight_layout()
plt.show()    
```
