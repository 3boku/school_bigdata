```python
import csv
import matplotlib.pyplot as plt

f = open('/weather.csv', encoding='euc-kr')
data = csv.reader(f)
header = next(data)
data = list(data)
f.close()
```

```python
high = []
for row in data:
  if row [0].split('-') == '08' and row[0].split('-')[2] == '15':
    if row[2] != '':
      high.append(float(row[2]))
      low.append(float(row[2]))
```

```python
avg = []
years = [1960, 1961, 1962]
sum = 0
days = 0

for row in data:
  if int(row[0].split('-')[0]) == years[-1]:
    if row[1] != '':
      sum += float(row[1])
      days += 1
  elif int(row[0].split('-')[0]) > years[-1]:
    avg.append(sum/days)
    if int(row[0].split('-')[0]) == 2021:
      break
    else:
      years.append(years[-1]+1)
      sum = float(row[1])
      days = 1

plt.figure(figsize = (8,3))
plt.title('Average temperture 1960~2020')
plt.xlabel('year')
plt.ylabel('temperature')
plt.plot(years, avg, color = 'red')
plt.show()
```

```python
result = []
for row in data:
  if row[2] != '':
    result.append(float(row[2]))
plt.hist(result, bins = 100)
plt.show()
```

```python
aug = []
jan = []

for row in data:
  month = row[0].split('-')[1]
  if row[2] != '':
    if month == '08':
      aug.append(float(row[2]))
    if month == '01':
      jan.append(float(row[2]))


plt.hist(aug, bins = 100, color = 'red', label='AUG')
plt.hist(jan, bins = 100, color = 'blue', label = 'JAN')
plt.legend()
plt.show()
```
